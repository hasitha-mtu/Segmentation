```python
model_result_path = "../output/17_07_2025/model_results.csv"
```

```python
df = pd.read_csv(model_result_path)
df
```

```python
# Normalize 'col2' to the range 0-1
column_to_normalize = 'Inference Time (ms)'

# Calculate min and max for the specific column
min_val = df[column_to_normalize].min()
max_val = df[column_to_normalize].max()

# Apply the min-max normalization formula
df[f'{column_to_normalize}_normalized'] = (df[column_to_normalize] - min_val) / (max_val - min_val)

print("\nDataFrame after normalizing 'col2' to 0-1:")
print(df)

# Normalize 'col3' to a different range, e.g., 0-10
column_to_normalize_2 = 'Inference Time (ms)'
min_new_range = 0
max_new_range = 10

min_val_2 = df[column_to_normalize_2].min()
max_val_2 = df[column_to_normalize_2].max()

df[f'{column_to_normalize_2}_normalized_0_10'] = (
    (df[column_to_normalize_2] - min_val_2) / (max_val_2 - min_val_2)
) * (max_new_range - min_new_range) + min_new_range

print("\nDataFrame after normalizing 'Inference Time (ms)' to 0-10:")
print(df)
```
