```python
def calculate_precision(ground_truth_mask, predicted_mask):
    """
    Calculates Precision for binary segmentation masks.

    Args:
        ground_truth_mask (np.array): A binary NumPy array representing the ground truth.
                                      (e.g., 1 for foreground/positive, 0 for background/negative)
        predicted_mask (np.array): A binary NumPy array representing the model's prediction.
                                   (e.g., 1 for foreground/positive, 0 for background/negative)

    Returns:
        float: The calculated Precision score. Returns 0.0 if TP + FP is 0
               (i.e., no positive predictions were made).
    """
    # Ensure masks are boolean or integer (0 or 1)
    if ground_truth_mask.dtype != bool:
        ground_truth_mask = ground_truth_mask.astype(bool)
    if predicted_mask.dtype != bool:
        predicted_mask = predicted_mask.astype(bool)

    # True Positives (TP): Pixels where prediction is 1 AND ground truth is 1
    true_positives = np.sum(np.logical_and(predicted_mask == 1, ground_truth_mask == 1))

    # False Positives (FP): Pixels where prediction is 1 BUT ground truth is 0
    false_positives = np.sum(np.logical_and(predicted_mask == 1, ground_truth_mask == 0))

    # Calculate Precision
    # Avoid division by zero if there are no positive predictions
    denominator = true_positives + false_positives
    if denominator == 0:
        precision = 0.0
    else:
        precision = true_positives / denominator

    print(f"  True Positives (TP): {true_positives}")
    print(f"  False Positives (FP): {false_positives}")
    print(f"  TP + FP (Total Positive Predictions): {denominator}")

    return precision
```

```python
import numpy as np

def calculate_precision_adapted(y_true, y_pred, evaluation_mask, threshold=0.5, positive_class_channel=1):
    """
    Calculates Precision for binary semantic segmentation, adapting to given shapes
    and using an evaluation mask.

    Args:
        y_true (np.array): Ground truth mask (512, 512, 1), containing 0 or 1.
        y_pred (np.array): Model prediction (512, 512, 3), likely probabilities.
                           Assumes one channel represents the positive class.
        evaluation_mask (np.array): Mask for evaluation (512, 512, 1), containing 0 or 1.
                                    Only pixels where this mask is 1 are considered.
        threshold (float): Threshold to binarize y_pred probabilities. Default is 0.5.
        positive_class_channel (int): The index of the channel in y_pred that corresponds
                                      to the positive class (e.g., flood). Default is 1.

    Returns:
        float: The calculated Precision score. Returns 0.0 if TP + FP is 0.
    """
    # 1. Squeeze singleton dimensions for y_true and evaluation_mask
    #    (512, 512, 1) -> (512, 512)
    y_true_squeezed = y_true.squeeze()
    evaluation_mask_squeezed = evaluation_mask.squeeze()

    # 2. Binarize y_pred to (512, 512)
    #    Select the channel for the positive class and apply threshold
    if y_pred.ndim == 3 and y_pred.shape[2] > positive_class_channel:
        # Assuming y_pred[:, :, positive_class_channel] gives probabilities for the positive class
        predicted_binary_mask = (y_pred[:, :, positive_class_channel] >= threshold).astype(bool)
    elif y_pred.ndim == 2:
        # If y_pred was already 2D (e.g., from a previous squeeze), just apply threshold
        predicted_binary_mask = (y_pred >= threshold).astype(bool)
    else:
        raise ValueError(f"y_pred shape {y_pred.shape} is not supported. Expected (H,W,C) or (H,W).")

    # Ensure all masks are boolean for logical operations
    y_true_bool = y_true_squeezed.astype(bool)
    predicted_bool = predicted_binary_mask.astype(bool)
    eval_mask_bool = evaluation_mask_squeezed.astype(bool)

    # 3. Apply the evaluation_mask before calculating TP and FP
    # Only consider pixels where evaluation_mask_squeezed is True (1)

    # True Positives (TP): Pixels correctly predicted as positive AND within the evaluation mask
    true_positives = np.sum(np.logical_and(predicted_bool, y_true_bool, eval_mask_bool))

    # False Positives (FP): Pixels incorrectly predicted as positive AND within the evaluation mask
    # Prediction is 1, Ground Truth is 0, AND within evaluation mask
    false_positives = np.sum(np.logical_and(predicted_bool, np.logical_not(y_true_bool), eval_mask_bool))

    # Calculate Precision
    denominator = true_positives + false_positives
    if denominator == 0:
        precision = 0.0
    else:
        precision = true_positives / denominator

    print(f"\n--- Calculation Details ---")
    print(f"  Shape of y_true (squeezed): {y_true_squeezed.shape}")
    print(f"  Shape of y_pred (binarized): {predicted_binary_mask.shape}")
    print(f"  Shape of evaluation_mask (squeezed): {evaluation_mask_squeezed.shape}")
    print(f"  True Positives (TP) within mask: {true_positives}")
    print(f"  False Positives (FP) within mask: {false_positives}")
    print(f"  Total Positive Predictions within mask (TP + FP): {denominator}")

    return precision

# --- Demonstration with example data ---

# Example data generation (replace with your actual data)
H, W = 512, 512

# Simulate y_true (e.g., a central square as flood)
y_true_example = np.zeros((H, W, 1))
y_true_example[H//4 : 3*H//4, W//4 : 3*W//4] = 1 # A square of 1s in the center

# Simulate y_pred (e.g., probabilities for 3 classes)
# Channel 0: background prob, Channel 1: flood prob, Channel 2: another class prob
y_pred_example = np.random.rand(H, W, 3) # Random probabilities initially

# Let's make flood prob higher in the true flood region, lower elsewhere
y_pred_example[H//4 : 3*H//4, W//4 : 3*W//4, 1] = np.random.uniform(0.6, 0.9, (H//2, W//2)) # High flood prob
y_pred_example[H//4 : 3*H//4, W//4 : 3*W//4, 0] = 0.1 # Low background prob
y_pred_example[H//4 : 3*H//4, W//4 : 3*W//4, 2] = 0.1 # Low other class prob

y_pred_example[:, :, 1] = np.where(y_true_example.squeeze() == 1,
                                   np.random.uniform(0.7, 0.9, (H, W)), # Higher prob for true flood
                                   np.random.uniform(0.1, 0.3, (H, W))) # Lower prob for true background

# Normalize probabilities to sum to 1 (optional, depends on model output)
y_pred_example = y_pred_example / np.sum(y_pred_example, axis=-1, keepdims=True)


# Simulate mask (e.g., only evaluate the top-left quadrant due to weak supervision confidence)
evaluation_mask_example = np.zeros((H, W, 1))
evaluation_mask_example[0 : H//2, 0 : W//2] = 1 # Top-left quadrant is 1

print("--- Calling calculate_precision_adapted ---")
precision_score = calculate_precision_adapted(
    y_true=y_true_example,
    y_pred=y_pred_example,
    evaluation_mask=evaluation_mask_example,
    threshold=0.5, # Adjust if your model's output needs a different threshold
    positive_class_channel=1 # Adjust based on which channel in y_pred is your positive class
)

print(f"\nCalculated Precision: {precision_score:.4f}")

# --- Example with a simplified y_pred directly as probabilities for the target class ---
print("\n--- Simplified Example: y_pred already target class probabilities ---")
# If your model output was directly probabilities for the positive class (e.g., from a softmax for 2 classes)
y_pred_simple_prob = y_pred_example[:, :, 1] # Just taking the positive class channel directly

# Make it 3D again to fit the function signature (though the function handles 2D too)
y_pred_simple_3d = np.expand_dims(y_pred_simple_prob, axis=-1)
y_pred_simple_3d = np.repeat(y_pred_simple_3d, 3, axis=-1) # Duplicate to 3 channels for example

precision_score_simple = calculate_precision_adapted(
    y_true=y_true_example,
    y_pred=y_pred_simple_3d, # Pass the 3-channel version
    evaluation_mask=evaluation_mask_example,
    threshold=0.5,
    positive_class_channel=0 # Now assume channel 0 is the positive class as we duplicated it
)
print(f"\nCalculated Precision (Simplified y_pred): {precision_score_simple:.4f}")
```
